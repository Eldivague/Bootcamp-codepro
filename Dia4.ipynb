```python
# challenge 2

# Diccionario
Tarjeta_personal = {}

# Bucle

while True:

    salir_del_bucle = input('Deseas salir?')
      
    if salir_del_bucle == 'si':
        break

    anhadir_clave = input('anhade o modifica una clave')

    anhadir_un_valor = input('anhade el valor')

    if 'si' == input('desea borrar una clave ?'):
        del Tarjeta_personal[input('Que dato desea borrar')]
    
    Tarjeta_personal[anhadir_clave] = anhadir_un_valor
    print(Tarjeta_personal)
```

```python
# challenge 3

Agenda_de_contactos = {}

while True:

    salir_del_bucle = input('Deseas salir?')
      
    if salir_del_bucle == 'si':
        break

    anhadir_contacto = input('Ingrea un contacto para agendar')

    anhadir_nro_del_contacto = input('ingresa el nro')

    if 'si' == input('desea borrar un contacto ?'):
        del Agenda_de_contactos[input('Que dato desea borrar')]
    
    Agenda_de_contactos[anhadir_contacto] = anhadir_nro_del_contacto
    print(Agenda_de_contactos)

```

```python
# challenge 5

class Loro:
    especie = 'ave'

    def __init__(self, nombre, color):
        self.nombre = nombre
        self.color = color
    
    def volar(self):
        print(f'{self.nombre} esta volando por el patio')

    def hablar(self):
        print(f'{self.nombre} esta saludando a las visitas')

    
Loro1= Loro('Panchito', 'verde')

Loro1.volar()
Loro1.hablar()
```

```python
# challenge 6

class AnimalesDeCasa:
    especie = 'domestico'

    def __init__(self, nombre, color, sonido):
        self.nombre = nombre
        self.color = color
        self.sonido = sonido

    def hablar(self):
        print(f'soy {self.nombre} y hago {self.sonido}')
        
    def correr(self):
        print(f'soy {self.nombre} y saludo a mi amo')

class Perro(AnimalesDeCasa):

    def pasear(self):
        print('me gusta salir a pasear con mi amo')

perro1 = Perro('Poncho', 'Negro de manchas', 'wuf wuf')

perro1.correr()
perro1.hablar()
```

```python
# challenge 7

class Villano:
    especialidad = 'malvado'

    def __init__(self, nombre, poderes, caracteristica):
        self.nombre = nombre
        self.poderes = poderes
        self.caracteristicas = caracteristica

    def presentacion(self):
        print(f'me llamo {self.nombre} y derroto a mis enemigos con mis poderes de {self.poderes}')

    def ataque(self):
        print(f'lanzo vehiculos con mis poderes de {self.poderes}')


class Heroe:
    especialidad = 'bondadoso'

    def __init__(self,nombre, poderes, caracteristica):
        self.nombre = nombre
        self.poderes = poderes
        self.caracteristicas = caracteristica

    def presentacion(self):
        print(f'Mi nombre es {self.nombre} y defiendo a mi pueblo con mis poderes de {self.poderes}')

    def ataque(self):
        print(f'neutralizo y hago volar a los malos con mis poderes de {self.poderes}')

supercucarachoman = Villano('Supercucarachoman', 'controlar cucarachas', 'Agresivo')
hombrepalomon = Heroe('Hombrepalomon', 'controlar palomas', 'Piadoso')
        
supercucarachoman.presentacion()
supercucarachoman.ataque()

hombrepalomon.presentacion()
hombrepalomon.ataque()
```
